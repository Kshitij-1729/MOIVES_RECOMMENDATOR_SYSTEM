```python
df['genres'][0]  #df is an object(string) , we need to convert genres col into list of dictionary and extract eg action only
```

```python
import ast   # using this lib to convert the object to list
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
#for eg
```

```python
#extracting only the genre(name) from the col
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 

df['genres']=df['genres'].apply(convert)
```

```python
df['genres']
```

```python
df['keywords'][0]
```

```python
# similarly extracting the keyword (name from each dict of the row) 
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 

df['keywords']=df['keywords'].apply(convert)
df['keywords'].head()
```

```python
df.head(3)
```

```python
df['cast'][0]
```

```python
#extracting the top three cast of the movies in the cast col
def convert(text):
    count=1
    L=[]
    for i in ast.literal_eval(text):
        if count>3:
            break
        else:
            L.append(i['name'])
        count+=1
    return L

df['cast']=df['cast'].apply(convert)
df['cast'].head()
```

```python
df['cast'] = df['cast'].apply(lambda x:x[0:3])  #ensuring that only three names are only there
```

```python
df['crew'][0]  #we only want director name from this
```

```python
# where ever the multiple dict there is job='Director' get the name of it
def convert(text):
    L=[]
    for i in ast.literal_eval(text):
        if i['job']=='Director':
            L.append(i['name'])
    return L

df['crew']=df['crew'].apply(convert)
df['crew'].head()
```

```python
df.sample(3)
```

```python
#removing spaces from each col (especially b/t two words)
def remove_spaces(list):
    L=[]
    for i in list:
        L.append(i.replace(" ",""))
    return L
```

```python
df['cast'] = df['cast'].apply(remove_spaces)
df['crew'] = df['crew'].apply(remove_spaces)
df['genres'] = df['genres'].apply(remove_spaces)
df['keywords'] = df['keywords'].apply(remove_spaces)
```

#### Doing Vectorization(converting the text into vector) using Bag of Words NLP algo

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

stop word means (are,to,the) , we dont want these words to be account in making the vectors

```python
vector = cv.fit_transform(df['tags']).toarray()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
w1=ps.stem('loving')
w2=ps.stem('loved')
w3=ps.stem('love')
print(w1,w2,w3)
```

```python
def stemming(text):
    L=[]
    for i in text.split():    #conveting the string into list
        L.append(ps.stem(i))
    return " ".join(L)

df['tags']=df['tags'].apply(stemming)
df['tags']
```

```python
vector = cv.fit_transform(df['tags']).toarray()
```

```python
vector
```

```python
list(cv.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity  #cosine_similarrity is inverse of cosine_distance
```

```python
similarity = cosine_similarity(vector)
```

```python
# main function that recommends moive : find the moive indx then then goes to the similarity array and the finds the top 5 similar moives index from it and then display that top 5 movies
def recommend(movie):
    movie_indx = df[df['title'] == movie].index[0]
    distance = similarity[movie_indx]
    movie_list = sorted(list(enumerate(distance)), reverse=True, key=lambda x: x[1])[1:6]
    for i in movie_list:
        print(df.iloc[i[0]].title)
```
